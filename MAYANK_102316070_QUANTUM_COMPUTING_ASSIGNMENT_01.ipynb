```python
def is_normalized(vector):
  magnitude = np.linalg.norm(vector)
  if math.isclose(magnitude,1.0):
    print("The vector is normlaized vector")
  else:
    print("The vector is not normlaized vector")


def np_dot_cross_product(vector1,vector2):
  dot_product = np.dot(vector1,vector2)
  cross_product = np.cross(vector1,vector2)

  return dot_product,cross_product


def unit_vector(vector1):
  magnitude = np.linalg.norm(vector1)
  if magnitude == 0:
        return "Zero vector has no unit vector"1

  unit_vector = vector1/magnitude

  return unit_vector

def read_value():
  n = int(input("enter the number of values in vector : "))
  elements = []

  for i in range(n):
    element = float(input("enter the value :"))
    elements.append(element)

  return np.array(elements)


A = read_value()
B = read_value()

print(A)
print(B)


is_normalized(A)
is_normalized(B)


print("the dot product and cross product : ",np_dot_cross_product(A,B))

print(unit_vector(A))
print(unit_vector(B))


```

```python
def read_matrix():
  m = int(input("enter the number of rows : "))
  n = int(input("enter the number of columns : "))

  matrix = []

  for i in range(m):
    row = []
    for j in range(n):
      value = float(input(f"Enter element [{i}][{j}]: "))
      row.append(value)
    matrix.append(row)

  return np.array(matrix)


def transpose_matrix(matrix):
  return matrix.T

def determinant_matrix(matrix):
  if matrix.shape[0] != matrix.shape[1]:
    return "Determinant is defined only for the square matrices"
  return np.linalg.det(matrix)

def rank_matrix(matrix):
  if matrix.shape[0] != matrix.shape[1]:
    return "Rank is defined only for the square matrices"
  return np.linalg.matrix_rank(matrix) #or np.trace(matrix)

def eigen_matrix(matrix):
  if matrix.shape[0] != matrix.shape[1]:
    return "Eigen values and eigen vectors are defined only for the square matrices"
  eigen_values,eigen_vectors = np.linalg.eig(matrix)
  return eigen_values,eigen_vectors

def inverse_matrix(matrix):
  if matrix.shape[0] != matrix.shape[1]:
    return "Inverse is defined only for the square matrices"
  if np.linalg.det(matrix) == 0:
      return "Matrix is singular; inverse does not exist"
  return np.linalg.inv(matrix)

A = read_matrix()

print("\nMatrix A:")
print(A)

print("\nTranspose:")
print(transpose_matrix(A))

print("\nDeterminant:")
print(determinant_matrix(A))

print("\nRank:")
print(rank_matrix(A))

print("\nTrace:")
print(rank_matrix(A))

print("\nEigenvalues and Eigenvectors:")
print(eigen_matrix(A))

print("\nInverse:")
print(inverse_matrix(A))


```
